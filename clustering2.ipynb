```python
def quantize_speeds(dataframe):
    """
    quantizes speeds for given dataframe to reduce computation

    returns: quantized dataframe
    """
    output_dataframe = pd.DataFrame()
    
    cols = dataframe.columns
    
    for col in cols:
        output_dataframe[col] = dataframe[col].round(decimals=3)
    
    return output_dataframe
        
```

```python
def get_positions(dataframe):
    """
    calculates positions for each vehicle for the trip

    returns: positions dataframe
    """
    output_dataframe = pd.DataFrame()
    
    cols = dataframe.columns
    
    for col in cols:
        output_dataframe[col] = dataframe[col].cumsum()
    
    return output_dataframe
        
```

```python
class Cluster:
    
    def quantize_speeds(self, dataframe):
        """
        quantizes speeds for given dataframe to reduce computation

        returns: quantized dataframe
        """
        output_dataframe = pd.DataFrame()

        cols = dataframe.columns

        for col in cols:
            output_dataframe[col] = dataframe[col].round(decimals=3)

        return output_dataframe
   
    def get_positions(self, dataframe):
        """
        calculates positions for each vehicle for the trip

        returns: positions dataframe
        """
        output_dataframe = pd.DataFrame()

        cols = dataframe.columns

        for col in cols:
            output_dataframe[col] = dataframe[col].cumsum()

        return output_dataframe


    def get_clusters(self, dataframe, duration, cluster_radius, test_car):
        """
        clustering algorithm for offloading decision making

        returns: optimal cluster formation for given test vehicle
        """
#         drop_car=self.test_car  #ya wala dekh
        pos_cluster = {test_car: {}}
        
        position_test = dataframe[test_car][duration] #test vehicle position for ending timestep
        
        x = dataframe.drop(columns=[test_car]) 

        all_pos_timestep = x.iloc[duration, :] #positions data for all other cars for ending timestep
                
        vehicle_names = x.columns
        
        for service_veh in range(len(all_pos_timestep)):
            diff = abs(all_pos_timestep[service_veh] - position_test)
            
            if diff < cluster_radius:
                if round(diff, 2) != 0.0:
                    pos_cluster[test_car][vehicle_names[service_veh]] = round(diff, 2)

        return pos_cluster
    
    def PBTO(self, distance, size):
        """
        computes time based on distance of service vehicle from test vehicle and the size of task in MBs

        returns: time for transmission, computation and retransmission
        """
        B = 10
        Pt = 0.5
        No = 10E-2
        S = Pt/(distance**2)
        SNR = S/No
        DR = B*np.log2(1+SNR)
        Txt = size/DR
        Cxt = 5
        Rxt = (size/5)/DR
        print(f'- Txt {Txt} - Cxt {Cxt} - Rxt {Rxt}')
        time = Txt + Cxt + Rxt
        
        
        return time    
    
    def selected_sorted_nearest_cars(self, vehicle):
    
        near_cars=get_clusters(fab_data,duration=1,cluster_radius=20,test_car=vehicle)
        sort_orders = sorted(near_cars[vehicle].items(), key=lambda x: x[1])

        names = []
        #select first four cars from list

        for i in range(4):
            name = sort_orders[i][0]
            names.append(name)
        return names
    

    def final_cars_availble(self, test_car, names, duration, dataframe):

        pos_cluster = {test_car: {}}

        position_test = dataframe[test_car][duration] #test vehicle position for ending timestep

        x = dataframe.drop(columns=[test_car]) 

        all_pos_timestep = x.iloc[duration, :] #positions data for all other cars for ending timestep

        vehicle_names = x.columns


        ending_distances = []
        times = []

        for service_veh in range(len(names)):
            diff = abs(all_pos_timestep[service_veh] - position_test)
            ending_distances.append(diff)

        for distance in ending_distances:
            times.append(self.PBTO(distance, 5))
        print("ending distances", ending_distances)
        mean = np.mean(times)

        print("\n\n",mean)
    

# acha wala
cluster_obj = Cluster()

quantized_df = cluster_obj.quantize_speeds(df)

positions_df = cluster_obj.get_positions(quantized_df)


clusters = cluster_obj.get_clusters(fab_data, duration=10, cluster_radius=50, test_car="vehicle_6")
# print(clusters)

distances = list(clusters[list(clusters.keys())[0]].values())

distances = sorted(distances)

if len(distances) >= 4:
    distances = distances[:4]

print(distances)
times = []

for distance in distances:
    times.append(cluster_obj.PBTO(distance,0.5))


mean = np.mean(times)
round(mean, 2)
# names = cluster_obj.selected_sorted_nearest_cars("vehicle_6")

# final_cars_availble('vehicle_6', names, 10, fab_data)
```
